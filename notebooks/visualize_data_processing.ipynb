```python
import torch
import matplotlib.pyplot as plt
%matplotlib inline

def visualize_image(tensor: torch.Tensor, index: int = 0) -> None:
    if tensor.ndim == 4:  # Batch of images
        image = tensor[index] 
    else:  # Single image
        image = tensor  
    
    if tensor.ndim == 4:  # If batch (B, C, H, W)
        image = tensor[index]  # Extract single image
    else:  # Else single image (C, H, W)
        image = tensor  

    # Move channels to matplotlib RGB format (H,W,C)
    image = image.permute(1, 2, 0).numpy()

    # Display the image
    plt.figure(figsize=(4, 4))
    plt.imshow(image)
    plt.axis("off")
    plt.show()
```

# Resize image

```python
visualize_image(resize)
# Can see image gets distorted
```

# Pad and Resize

```python
visualize_image(pad_resize)
```

# Resize and pad

```python
visualize_image(resize_pad)
```
